### 2. numeric한 독립변수들

우선적으로 쉽게 할 수 있는 것은 numeric 변수들을 파악하는 것이다
#### - 상관관계 파악하기
종속변수인 saleprice와 상관관계가 있는 변수들을 찾아보기

```r
# 우선 numeric한 변수들을 따로 모아보자
numeric_vars = which(sapply(all, is.numeric))
numeric_names <- names(numeric_vars)
cat('There are', length(numeric_vars), 'numeric variables')
```

```r
all_numvar <- all[, numeric_vars]
cor_numvar <- cor(all_numvar, use="pairwise.complete.obs") #correlations of all numeric variables

#sort on decreasing correlations with SalePrice
cor_sorted <- as.matrix(sort(cor_numvar[,'SalePrice'], decreasing = TRUE))
print(cor_sorted)
cor_numvar
#select only high corelations
cor_high <- names(which(apply(cor_sorted, 1, function(x) abs(x)>0.5)))
cor_numvar <- cor_numvar[cor_high, cor_high]
print(cor_high)
corrplot.mixed(cor_numvar, tl.col="black", tl.pos = "lt")
```

총 10개의 변수가 0.5이상, 즉 집값과 상관성이 있었다.

따라서 10개의 변수들의 특징을 보다 자세하기 파악해 볼 필요성이 있다.

34개의 변수에서 결측치가 존재하네요, 결측치 너무 많아요... 이거 다 어떻게 채울까요...

#### PoolQC
풀의 퀄리티! , 결측치들은 풀이 없기 떄문일거에요  
 Ex   Excellent  
   Gd   Good  
   TA   Average/Typical  
   Fa   Fair  
   NA   No Pool  

```r
all$PoolQC[is.na(all$PoolQC)] <- 'None'
qualities <- c('None' = 0, 'Po' = 1, 'Fa' = 2, 'TA' = 3, 'Gd' = 4, 'Ex' = 5)
```

revalue 라는 펑션을 활용해서 쉽게 팩터형 변수를 수정할거에요

```r
all$PoolQC<-as.integer(revalue(all$PoolQC, qualities))
table(all$PoolQC)
```

#### MiscFeature
   Elev Elevator  
   Gar2 2nd Garage (if not described in garage section)  
   Othr Other  
   Shed Shed (over 100 SF)  
   TenC Tennis Court  
   NA   None  

```r
all$MiscFeature[is.na(all$MiscFeature)] <- 'None'
all$MiscFeature <- as.factor(all$MiscFeature)

ggplot(all[!is.na(all$SalePrice),], aes(x=MiscFeature, y=SalePrice)) +
        geom_bar(stat='summary', fun.y = "median", fill='blue') +
        scale_y_continuous(breaks= seq(0, 800000, by=100000), labels = comma) +
        geom_label(stat = "count", aes(label = ..count.., y = ..count..))
```
